```python
age = int(input("年齢は？"))

if age >= 18:
    student = input("学生ですか？(yes\no)")

    student = student.strip().lower()

    if student == "yes":
        print("学生対象です")
    else:
        print("通常料金です")

else:
    print("未成年です")
```

```python
age = int(input("年齢は？"))

if age >= 18:
    student = input("学生ですか？(yes\no)")
    student = student.strip().lower()

    if student == "yes":
        print("学生対象です")

    elif student == "no":
        print("通常料金です")

    else:
        print("yesかnoで答えてください")

else:
    print("未成年です")
```
